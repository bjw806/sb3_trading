```python
from openfe import OpenFE, transform, tree_to_formula
import pandas as pd
```

```python
df = pd.read_pickle("./data/train/day/2022/1/2022_01_01_1m.pkl")
```

```python

df['position'] = 'Hold'

df.loc['2021-12-31 16:00:00':'2021-12-31 16:59:00', 'position'] = 'Short'
df.loc['2021-12-31 20:00:00':'2021-12-31 21:10:00', 'position'] = 'Long'
df.loc['2022-01-01 05:00:00':'2022-01-01 05:30:00', 'position'] = 'Short'
```

```python

ofe = OpenFE()
features = ofe.fit(data=df, label=df['position'])
```
